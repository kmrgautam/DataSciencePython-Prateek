## Joining two datsets

There are various ways of joining two tables

## Syntax for joining.

First_dataset.merge(Second_dataset,how ='inner/left/right',on='Which key to join?')

## Inner join

```python

ord_dtls=ecom.merge(cust, how='inner', on='Order ID')
```

## Left join

Using Left joins


```python
# here cust is the first table and ecom is the 2nd table
cust_ord_1=cust.merge(ecom, how='left', on='Order ID')
```

```python
cust_ord_1.head(4)
```

```python
#in left join, we will have the same number of rows in the left table
cust_ord_1.shape
```

## Right joins

```python
# here cust is the first table and ecom is the 2nd table
ecom_orders_cust=cust.merge(ecom, how='right', on='Order ID')
```

```python
ecom_orders_cust.head(4)
```

```python
#in right join, we will have all the rows in the right table
ecom_orders_cust.shape
```

## Grouping

```python
# get the median age of all the passengers
comb['Age'].median()
```

```python
# get the median of age by the values of gender of the passengers
comb.groupby('Sex').Age.median()
```

```python
# get the median of fare by the values of Pclass of the passengers
comb.groupby(['Pclass']).Fare.median()
```

```python
# get the median of age by the values of Pclass of the passengers
comb.groupby(['Pclass']).Age.median()
```

```python
# get the median of fare and age by the values of Pclass of the passengers
comb.groupby(['Pclass'])['Fare','Age'].median()
```

```python
##We can use different aggregate functions in group by
# if we need the mean of fare and the median of age
comb.groupby(['Pclass']).agg({'Fare' : 'mean', 'Age' : 'median'})
```

## Crosstabs

```python

# Crosstabulation of sex and Passenger class
pd.crosstab(comb.Sex, comb.Pclass)
```

```python
# To get a chart of the crosstab
pd.crosstab(comb.Sex, comb.Pclass).plot(kind='bar');
```

## Pivot tables

Like in excel, we have pivots in pandas too

index=rows

columns=variables

values- the variable whose aggregation we want

aggfunc - the kind of agg func

```python
#Suppose we need a  pivot table of total age of passenger classes by their sex
comb.pivot_table(index='Sex',columns = 'Pclass',values='Age', aggfunc='sum')
```

```python
#Suppose we need a  pivot table of mean age of passenger classes by their sex
comb.pivot_table(index='Sex',columns = 'Pclass',values='Age', aggfunc='mean')
```

## Missing value treatment

```python
# in  your dataframe, use .info() to detect missing values (if any) 
comb.info()
```

```python
# extract all rows with Embarked as Null
comb[comb.Embarked.isnull()]
```

```python
# how many people embarked at different points C = Cherbourg, Q = Queenstown, S = Southampton
comb.Embarked.value_counts()
```

```python
# Which passengers who embarked from any of these 3 points have higher survival count
pd.crosstab(comb[comb.Survived != 999].Survived, comb[comb.Survived != 999].Embarked)
```

```python
# impute the missing values 
comb.loc[comb.Embarked.isnull(), 'Embarked']
#  if you want to put S in place of the missing value
#df.Embarked.fillna('S', inplace=True)
```

```python
# 2nd method : Get the fare of each class for each embarkment point
comb.groupby(['Pclass', 'Embarked']).Fare.median()
```

```python
# replace the missing values with 'C'
comb.Embarked.fillna('C', inplace=True)
```

```python
# check if any null value remaining
comb[comb.Embarked.isnull()]
```
