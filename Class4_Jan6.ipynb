# DATA FRAMES
Data frames are 2d data structures 

 Data is aligned in a tabular format.
 
 It consists of rows and columns 
 
 DataFrames can hold multiple data type
 
 
  The basic form is pd.DataFrames(data=,index=,dtype=,copy) 

```python
# Dataframes
# Default is pandas.DataFrame( data, index, columns, dtype, copy)
d1=pd.DataFrame()
print(d1)
```

```python
# Create a DataFrame from Lists
df=pd.DataFrame([1,2,3,4])
df
```

```python
# passing the column index
df1=pd.DataFrame([1,2,3,4],index=['a','d','c','z'],columns=['one'])
df1
```

```python

# Passing the datatype
df2=pd.DataFrame([1,2,3,4],index=['a','d','c','z'],columns=['one'],dtype='float64')
```

```python
# using a dictionary to create a DatFrame

df = pd.DataFrame({'age':[30, 2, 12, 4, 32, 33, 69],
                   'color':['blue', 'green', 'red', 'white', 'gray', 'black', 'red'],
                   'food':['Steak', 'Lamb', 'Mango', 'Apple', 'Cheese', 'Melon', 'Beans'],
                   'height':[165, 70, 120, 80, 180, 172, 150],
                   'score':[4.6, 8.3, 9.0, 3.3, 1.8, 9.5, 2.2],
                   'state':['NY', 'TX', 'FL', 'AL', 'AK', 'TX', 'TX']
                   },
                  index=['Jane', 'Nick', 'Aaron', 'Penelope', 'Dean', 'Christina', 'Cornelia'])

df
```
