Some terminology associated with Matplotlib

## Figure
The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures

## Axes
Most plotting ocurs on an Axes.
The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

```python
# We can also plot 2 graphs
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes using r for color, s for square;, g for green, ^ for triangles
plt.plot(t, t**2, 'rs', t, t**3, 'g^')
```

```python
# we can also make 2 plots by using two plot instances
plt.plot(t, t**2, 'rs')
plt.plot(t, t**3, 'g^')
# If we dont want to use the grid, then make grid=False, default is True
plt.grid('False')
```

```python
# Suppose we want to limit the axis
plt.plot(t, t**2, 'rs', t, t**3, 'g^')
plt.axis([1,5,2,6])
```

```python
# Another way to limit the axis is to use the xlim and ylim

plt.plot(t, t**2, 'rs', t, t**3, 'g^')
plt.xlim(-1,5)
plt.ylim(-1,10)
```
